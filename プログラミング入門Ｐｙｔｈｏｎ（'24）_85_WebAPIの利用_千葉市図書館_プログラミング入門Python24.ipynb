# 8.5 WebAPIの利用 千葉市図書館

## Excel形式のファイルの利用 千葉市図書館の来館者数

Excel形式のファイルを Pandas で分析する。

1. まず、千葉市図書館利用統計を開く。（収録時点では、以下のものになっている。 https://www.city.chiba.jp/kyoiku/shogaigakushu/chuotoshokan/kanri/toukeiopendate.html ）
    
2. 下の方の「千葉市図書館の来館者数の推移（平成２２年度から令和４年度）（エクセル：11KB）」を探す。（年度の部分は、授業収録後に更新されている可能性があります。そのときは最新版に読み替えてください。）

3. その部分のURLをブラウザでコピーする。（右クリックメニューで、「リンクのアドレスをコピー」）をする。（収録時点では、以下のものになっている。
    https://www.city.chiba.jp/kyoiku/shogaigakushu/chuotoshokan/kanri/documents/raikan.xlsx ）

3. 次のプログラムの 2 行目の the_url の値に、先程のURLを設定する。そして、
 4 行目から 8 行目を実行してみて、中を見て、読み飛ばす行や列を検討する。

【プログラム8501-1】

```
import pandas as pd # データフレームの解析
the_url = 'https://www.city.chiba.jp/kyoiku/shogaigakushu/chuotoshokan/kanri/documents/raikan.xlsx'

df = pd.read_excel(
    the_url
)

df # 中を見て、読み飛ばす行や列を検討する。
```

* 中を見て、読み飛ばす行や列を検討する。
    * 1 行目（来館者数の推移）を読み飛ばす
    * 1 列目（図書館名）を見出しにする。

そこで、オプションを付ける。

* 1 行目は不要。
  * 0 行目を読み飛ばす（最初の行が0行目になっている。）
  *  skiprows = [0]

* 1列目をデータのインデックスにする。
  * 0 列目をインデックスに指定。
  * index_col = 0

  以上のことを反映させて、再度読み込んで確認する。

【プログラム8501-2】

```
df = pd.read_excel(
    the_url,
    skiprows = [0],
    index_col = 0
)

df # 中を見て、読み飛ばす行や列を確認する。
```

元のデータは、各行が図書館名で、各列が年度であったので、 転置（元のデータの行と列を入れ替える）を行っておく必要がある。

【プログラム8501-3】

```
df = pd.read_excel(
    the_url,
    skiprows = [0],
    index_col = 0
)

df = df.T # 転置（元のデータの行と列を入れ替える）

df # 中を見て、グラフを描くデータフレームになっていることを確認する。
```

確認できたら、グラフを描いてみる。

【プログラム8501-4】

```
import matplotlib.pyplot as plt
# japanize-matplotlib は、matplotlib のグラフで日本語の文字を使うためのものである。
!pip install japanize-matplotlib
import japanize_matplotlib

df = pd.read_excel(
    the_url,
    skiprows = [0],
    index_col = 0
)

df = df.T # 転置（元のデータの行と列を入れ替える）
df.plot(figsize=(12,12), grid=True)
```
